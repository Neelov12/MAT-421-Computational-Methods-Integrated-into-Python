# 21.1: Numerical Integration Problem Statement

In this section, I create a python program that demonstrates the concept of this section

<bv>
First, an overview:
**Numerical Integration Problem Statement** is when we split $f(x)$ into a numerical grid $x$ consisting of $n+1$ points with spacing $h = \frac{b-a}{n}$

<bv>
In the code below, I split $f(x) = sin(x)$ into a numerical grid with 50 points

```python
import numpy as np
import matplotlib.pyplot as plt

def split_function(func, x_min, x_max, num_points):
    """Splits a function into a numerical grid with given x points."""
    x_values = np.linspace(x_min, x_max, num_points)
    y_values = func(x_values)
    return x_values, y_values

# Example function
def example_function(x):
    return np.sin(x)

# Define grid parameters
x_min, x_max = 0, 2 * np.pi  # Range
num_points = 50  # Number of points in the grid

# Get grid values
x_grid, y_grid = split_function(example_function, x_min, x_max, num_points)

# Plot the function
plt.plot(x_grid, y_grid, marker='o', linestyle='-', label='f(x) = sin(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Numerical Grid of Function')
plt.legend()
plt.grid(True)
plt.show()
```

# 21.2: Reimanns Integral

In this section, I write a python program that demonstrates the concept of this section.

<bv> An overview of **Reimanns Integral**:

<bv>
Riemann's integral is a fundamental concept in calculus that defines the integral of a function as the limit of Riemann sums. It approximates the area under a curve by dividing it into small subintervals, computing the function’s value at sample points, multiplying by the subinterval width, and summing these values. As the subintervals become infinitely small, the Riemann sum approaches the exact integral.



In the code below, I use this concept to approximate the area below $f(x) = sin(x)$ from $[0, π]$

```python
def riemann_integral(func, x_min, x_max, num_rects):
    """Approximates the integral using Riemann sums."""
    x_values = np.linspace(x_min, x_max, num_rects + 1)
    dx = (x_max - x_min) / num_rects
    midpoints = x_values[:-1] + dx / 2  # Midpoint rule
    y_values = func(midpoints)
    integral_approx = np.sum(y_values * dx)
    return x_values, y_values, integral_approx

# Example function
def example_function(x):
    return np.sin(x)

# Define grid parameters
x_min, x_max = 0, np.pi  # Range
num_rects = 10  # Number of rectangles

# Get Riemann sum approximation
x_grid, y_grid, integral_approx = riemann_integral(example_function, x_min, x_max, num_rects)

# Plot the function and Riemann sum rectangles
x_plot = np.linspace(x_min, x_max, 1000)
y_plot = example_function(x_plot)
plt.plot(x_plot, y_plot, label='f(x) = sin(x)', color='blue')
plt.bar(x_grid[:-1] + (x_max - x_min) / (2 * num_rects), y_grid, width=(x_max - x_min) / num_rects, alpha=0.3, color='red', align='center', edgecolor='black', label='Riemann Sum')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Riemann Integral Approximation')
plt.legend()
plt.grid(True)
plt.show()

print(f"Approximate Integral: {integral_approx}")
```

# 21.3: Trapezoid Rule

In this section, I write a python program that demonstrates the concept of this section.

<bv> An overview of **Trapezoid Rule**:

<bv> The trapezoid rule is a numerical integration method that approximates the integral of a function by dividing the area under the curve into trapezoids rather than rectangles, improving the approximation power from Reimanns Integral

```python
def trapezoid_rule(func, x_min, x_max, num_trapezoids):
    """Approximates the integral using the trapezoid rule."""
    x_values = np.linspace(x_min, x_max, num_trapezoids + 1)
    y_values = func(x_values)
    dx = (x_max - x_min) / num_trapezoids
    integral_approx = (dx / 2) * (y_values[0] + 2 * np.sum(y_values[1:-1]) + y_values[-1])
    return x_values, y_values, integral_approx

# Example function
def example_function(x):
    return np.sin(x)

# Define grid parameters
x_min, x_max = 0, np.pi  # Range
num_trapezoids = 10  # Number of trapezoids

# Get trapezoid rule approximation
x_grid, y_grid, integral_approx = trapezoid_rule(example_function, x_min, x_max, num_trapezoids)

# Plot the function and trapezoidal approximation
x_plot = np.linspace(x_min, x_max, 1000)
y_plot = example_function(x_plot)
plt.plot(x_plot, y_plot, label='f(x) = sin(x)', color='blue')
plt.fill_between(x_grid, y_grid, alpha=0.3, color='red', label='Trapezoidal Approximation')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Trapezoid Rule Approximation')
plt.legend()
plt.grid(True)
plt.show()

print(f"Approximate Integral: {integral_approx}")
```
