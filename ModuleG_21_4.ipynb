# Module G: 21.4 Simpson's Rule

In this section, I will write a python program that demonstrates the concept of Simpson's Rule

First, an overview of Simpson's Rule:

<bv>

Simpsonâ€™s Rule is a numerical method for estimating the definite integral of a function. It works by approximating the area under a curve using a combination of parabolas, rather than straight lines (as in the Trapezoidal Rule).

It is most effective when the function being integrated is smooth and can be closely matched by a quadratic function over small intervals.

This program demonstrates this concept by approximating the integral of $f(x) = cos(x) $

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Define the function to integrate
def f(x):
    return np.cos(x)  # You can change this to any smooth function

# Simpson's Rule implementation
def simpsons_rule(f, a, b, n):
    if n % 2 != 0:
        raise ValueError("n must be even")

    h = (b - a) / n
    x = np.linspace(a, b, n + 1)
    y = f(x)

    S = y[0] + y[-1] + 4 * np.sum(y[1:-1:2]) + 2 * np.sum(y[2:-2:2])
    return (h / 3) * S

# Parameters
a = 0
b = np.pi
n = 10  # Must be even

# Compute approximate and exact values
approx = simpsons_rule(f, a, b, n)
exact, _ = quad(f, a, b)

# Print results
print(f"Simpson's Rule Approximation: {approx}")
print(f"Exact Integral Value:         {exact}")
print(f"Absolute Error:               {abs(approx - exact)}")

# Plotting the function and points
x_vals = np.linspace(a, b, 1000)
y_vals = f(x_vals)
x_nodes = np.linspace(a, b, n + 1)
y_nodes = f(x_nodes)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='f(x)', color='blue')
plt.plot(x_nodes, y_nodes, 'ro', label='Sample Points')
plt.title("Simpson's Rule Approximation")
plt.fill_between(x_vals, y_vals, alpha=0.2, color='orange', label='Area Under Curve')
plt.legend()
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```
